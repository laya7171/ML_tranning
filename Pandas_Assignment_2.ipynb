## Merge , Join and Concatenate

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/LOTR.csv")
```

```python
df
```

```python
df2=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/LOTR 2.csv")
df2
```

```python
df
```

```python
pd.concat([df,df2])
```

```python
#Inner Join
df.merge(df2, how= 'inner', on=['FellowshipID','FirstName'])
```

```python
#Outer Join
df.merge(df2, how= 'outer')
```

```python
#Left Join
df.merge(df2, how= 'left')
```

```python
#Right Join
df.merge(df2, how= 'right')
```
